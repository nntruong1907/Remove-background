```python
from rembg import remove
from PIL import Image
import os

def remove_bg(raw_folder):
    print("raw_folder: ", raw_folder)
    for image_name in os.listdir(raw_folder):
        print("image: ", image_name)
        image_path = os.path.join(raw_folder, image_name)
        print("image_path: ", image_path)
        
        img, _ = image_name.split(".")
        out_image = img + "-removebg.png" 
        out_path = os.path.join(raw_folder, out_image)
        print("out_path: ", out_path)
        
        input = Image.open(image_path)
        
        output = remove(input)
        output.save(out_path)
        
        os.remove(image_path)
        
remove_bg("./BMW")

#clcoding.com
```
